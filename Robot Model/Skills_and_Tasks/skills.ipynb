## Skills Specification
24 Skills at 8 Levels


*   Piece connecting Skills
    *  (Skill 1) How to Connect two pieces together

*   Piece Picking Skills
    *  (Skill 2.1) When a Lamp should be used
    *  (Skill 2.2) When a Motor should be used
    *  (Skill 2.3) When a Switch should be used
    *  (Skill 2.4) When a Button should be used  
    *  (Skill 2.5) When an LED should be used
    *  (Skill 2.6) When a Speaker should be used
    *  (Skill 2.7) When a Buzzer should be used
    *  (Skill 2.8) When an IC-Music should be used
    *  (Skill 2.9) When a FM Radio receiver should be used

*   Closed Circuit Skills
    *  (Skill 3) That the circuit should be a closed circuit

*   LED Skills
    *  (Skill 4.1) The correct directionality of LED
    *  (Skill 4.2) When an LED is placed on the board, a resistor is needed

*   Resistor Skills
    *  (Skill 5.1) How a resistor can be used to lower the light or music
    *  (Skill 5.2) How to short a resistor

*   Music Circuit Skills
    *  (Skill 6.1) Power MC - How to connect the Music Circuit (MC) to power. (The positive of the MC should be connected to the positive of the battery, and the negative of the MC should be connected to the negative of the battery.)
    *  (Skill 6.2) Signal MC - How to connect the positive of the battery to the signal of the MC (enables music to be played in a rhythm).
    *  (Skill 6.3) Trigger MC - How to correctly connect the positive side of the battery to the Trigger input on the MC.
    *  (Skill 6.4) Hold MC - How to correctly connect the positive side of the battery to the Hold input on the MC.

*   FM Skills
    *  (Skill 7.1) How to power an FM
    *  (Skill 7.2) How to change the signal on the FM

*   Gate Skills
    *  (Skill 8.1) How to create an AND gate (Used for tasks that require both a switch to be on and a button pressed for the circuit to work.)
    *  (Skill 8.2) How to create an OR gate (Used for tasks that require either a switch to be on or a button pressed for the circuit to work.)
    *  (Skill 8.3) How to create an NOT gate (Used for tasks that require the switch to be off for the circuit to work.)

## Task Specification
40 Tasks at 6 Levels

Level 1.a - Baseline w/ lamp
(Skills used: When to use lamp, When to use Switch/Button, Connect Lamp to Switch/Button, Closed Circuit)
- (Task 1.a.1 - Lamp) Build a circuit that has a constant lamp on
- (Task 1.a.2 - Switch, Lamp) Build a circuit that you can turn a lamp on and off using a switch
- (Task 1.a.3 - Button, Lamp) Build a circuit that you can turn a lamp on and off using a button

Level 1.b - Baseline w/ motor
(Skills used: When to use motor, When to use Switch/Button, Connect motor to Switch/Button, Closed Circuit)
- (Task 1.b.1 - Motor) Build a circuit that constantly will have a motor spinning
- (Task 1.b.2 - Switch, Motor) Build a circuit that you can turn a motor on and off using a switch
- (Task 1.b.3 - Button, Motor) Build a circuit that you can turn a motor on and off using a button

Level 2 - LED
(Skills used: When to use LED, Use Resistor along with LED, Direction of LED)
- (Task 2.0.1 - LED) Build a circuit that has a constant LED on
- (Task 2.0.2 - Switch, LED) Build a circuit that you can turn a LED on and off using a switch

Level 3 - FM
(Skills used: When to use FM, When to use Speaker, When to use a Buzzer)
- (Task 3.0.1 - FM, Speaker) Build a circuit that that you can listen to the FM radio via the Speaker
- (Task 3.0.2 - FM, Buzzer) Build a circuit that that you can listen to the FM radio via the Buzzer
- (Task 3.0.3 - FM, Speaker, Switch) Build a circuit that you can listen to the FM radio via the Speaker when a switch is turned on
- (Task 3.0.4 - FM, Speaker, Button) Build a circuit that you can listen to the FM radio via the Speaker when a button is pressed
- (Task 3.0.5 - FM, Buzzer, Switch) Build a circuit that you can listen to the FM radio via the Buzzer when a switch is turned on
- (Task 3.0.6 - FM, Buzzer, Button) Build a circuit that you can listen to the FM radio via the Buzzer when a button is pressed

Level 4 - Bright and Dim Light
(Skills used: When to use Lamp, When to use Resistor, How a resistor can be used to lower the light or music, How to short a resistor)
- (Task 4.0.1 - Lamp, Resistor, Switch) Build a circuit that has a bright light when a switch is turned on and a dim light when a switch is turned off
- (Task 4.0.2 - Lamp, Resistor, Button) Build a circuit that has a bright light when a button is pressed and a dim light when a button is not pressed

Level 5.1 - Music Circuit
(Skills used: When to use Music Circuit, How to connect Power/Signal/Trigger/Hold, Connect Speaker to Music Circuit)
- (Task 5.1.1 - MC, Speaker) Build a circuit that plays music
- (Task 5.1.2 - MC, Speaker, Switch) Build a circuit that plays music when a switch is turned on
- (Task 5.1.3 - MC, Speaker, Button) Build a circuit that plays music when a button is pressed

Level 5.2 - Music Circuit w/ resistor
(Skills used: How a resistor can be used to lower the light or music, How to connect resistor to Music Circuit)
- (Task 5.2.1 - MC, Speaker, Resistor) Build a circuit that plays low sounding music.
- (Task 5.2.2 - MC, Speaker, Resistor, Switch) Build a circuit that plays low sounding music when a switch is turned on.
- (Task 5.2.3 - MC, Speaker, Resistor, Button) Build a circuit that plays low sounding music when a button is pressed.

Level 5.3 - Music Circuit w/ LED
(Skills used: When to use LED, Use Resistor along with LED, Direction of LED, Connect LED to Music Circuit)
- (Task 5.3.1 - MC, Speaker, LED) Build a circuit that blinks a light in the rythm of a song
- (Task 5.3.2 - MC, Speaker, LED, Switch) Build a circuit that blinks a light in the rythm of a song when a switch is turned on
- (Task 5.3.3 - MC, Speaker, LED, Button) Build a circuit that blinks a light in the rythm of a song when a button is pressed

Level 5.4 - Music Circuit w/ Lamp & Resistor
(Skills used: When to use Lamp, When to use Resistor, Connect Lamp & Resistor to Music Circuit)
- (Task 5.4.1 - MC, Speaker, Lamp, Resistor) Build a circuit that blinks a low light in the rhythm of a song.
- (Task 5.4.2 - MC, Speaker, Lamp, Resistor, Switch) Build a circuit that blinks a low light in the rhythm of a song when a switch is turned on.
- (Task 5.4.3 - MC, Speaker, Lamp, Resistor, Button) Build a circuit that blinks a low light in the rhythm of a song when a button is pressed.

Level 6.1.a - Gate skills w/ Lamp
(Skills used: When to use AND/OR gate)
- (Task 6.1.a.1 - AND, Lamp, Switch, Button) Build a circuit that turns on a light when a switch is turned on AND a button is pressed.
- (Task 6.1.a.2 - OR, Lamp, Switch, Button) Build a circuit that turns on a light when a switch is turned on OR a button is pressed.
- (Task 6.1.a.3 - NOT, Lamp, Switch) Build a circuit that turns on a light when the switch is turned off but does not turn on a light when the switch is turned on

Level 6.1.b - Gate skills w/ Motor
- (Task 6.1.b.1 - AND, Motor, Switch, Button) Build a circuit that spins a motor when a switch is turned on AND a button is pressed.
- (Task 6.1.b.2 - OR, Motor, Switch, Button) Build a circuit that spins a motor when a switch is turned on OR a button is pressed.
- (Task 6.1.b.3 - NOT, Motor, Switch) Build a circuit that spins a motor when the switch is turned off but does not spin a motor when the switch is turned on

Level 6.2 - Gate skills w/ Music Circuit
- (Task 6.2.1 - AND, MC, Speaker,Switch, Button) Build a circuit that plays music when a switch is turned on AND a button is pressed.
- (Task 6.2.2 - OR, MC, Speaker, Switch, Button) Build a circuit that plays music when a switch is turned on OR a button is pressed.
- (Task 6.2.3 - NOT, MC, Switch) Build a circuit that plays music when the switch is turned off but does not play music when the switch is turned on

Level 6.3 - Gate skills w/ Music Circuit & LED
- (Task 6.3.1 - AND, LED, MC, Speaker, Switch, Button) Build a circuit that blinks a light in a rhythm of a song when a switch is turned on AND a button is pressed.
- (Task 6.3.2 - OR, LED, MC, Speaker, Switch, Button) Build a circuit that blinks a light in a rhythm of a song when a switch is turned on OR a button is pressed.
- (Task 6.3.3 - NOT, LED, MC, Speaker, Switch) Build a circuit that blinks a light in a rhythm of a song when the switch is turned off but does not blink a light in a rhythm of a song when the switch is turned on
